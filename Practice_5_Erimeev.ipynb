```python
#1 задание.
from sklearn.linear_model import LinearRegression

# Используем все входные переменные
X_all = df[['TV', 'radio', 'newspaper']]
y = df['sales']

# Разделяем данные на обучающую и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(X_all, y, random_state=1)

# Обучаем модель линейной регрессии
lr_all = LinearRegression()
lr_all.fit(X_train, y_train)

# Выводим коэффициенты
print("значение свободного члена(пересечение):", lr_all.intercept_)
print("Коэффициенты", lr_all.coef_)

# Предсказание и расчет MSE
y_pred_all = lr_all.predict(X_test)
mse_all = mean_squared_error(y_test, y_pred_all)
print("MSE:", mse_all)
```

```python
#2 задание.
from sklearn.preprocessing import PolynomialFeatures

# Исследуем зависимости MSE от степени полинома
max_pow = 10
mse_values = []

for degree in range(1, max_pow + 1):
    poly = PolynomialFeatures(degree)
    X_poly = poly.fit_transform(X_all)

    X_train_poly, X_test_poly, y_train_poly, y_test_poly = train_test_split(X_poly, y, random_state=1)
    lr_poly = LinearRegression()
    lr_poly.fit(X_train_poly, y_train_poly)

    y_pred_poly = lr_poly.predict(X_test_poly)
    mse = mean_squared_error(y_test_poly, y_pred_poly)
    mse_values.append(mse)

# Визуализируем зависимости MSE от степени полинома
plt.plot(range(1, max_pow + 1), mse_values, marker='o')
plt.title('MSE в зависимости от степени полинома')
plt.xlabel('Степень полинома')
plt.ylabel('MSE')
plt.show()

```

```python
#3 задание.
from sklearn.linear_model import Ridge
from sklearn.preprocessing import StandardScaler

# Нормализация признаков
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_poly)
X_test_scaled = scaler.transform(X_test_poly)

# Исследуем зависимость MSE от alpha
alpha_values = np.logspace(-3, 3, 100)
mse_ridge_values = []
max_coef_values = []

for alpha in alpha_values:
    ridge = Ridge(alpha=alpha)
    ridge.fit(X_train_scaled, y_train_poly)

    y_pred_ridge = ridge.predict(X_test_scaled)
    mse_ridge = mean_squared_error(y_test_poly, y_pred_ridge)

    mse_ridge_values.append(mse_ridge)
    max_coef_values.append(np.max(np.abs(ridge.coef_)))

# Визуализируем зависимости
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(alpha_values, mse_ridge_values, marker='o')
plt.title('MSE в зависимости от alpha(Гребневая)')
plt.xscale('log')
plt.xlabel('Alpha')
plt.ylabel('MSE')

plt.subplot(1, 2, 2)
plt.plot(alpha_values, max_coef_values, marker='o')
plt.title('Максимальный коэффициент в зависимости от alpha(Гребневая)')
plt.xscale('log')
plt.xlabel('Alpha')
plt.ylabel('Максимальный коэффициент')

plt.show()

```

```python
#4 задание.
from sklearn.linear_model import Lasso
from sklearn.preprocessing import StandardScaler

# Нормализация признаков
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_poly)
X_test_scaled = scaler.transform(X_test_poly)

# Исследуем зависимость MSE от alpha
mse_lasso_values = []
max_coef_lasso_values = []

alpha_values = np.logspace(-3, 3, 100)
for alpha in alpha_values:
    lasso = Lasso(alpha=alpha, max_iter=20000)  # Увеличиваем количество итераций
    lasso.fit(X_train_scaled, y_train_poly)

    y_pred_lasso = lasso.predict(X_test_scaled)
    mse_lasso = mean_squared_error(y_test_poly, y_pred_lasso)

    mse_lasso_values.append(mse_lasso)
    max_coef_lasso_values.append(np.max(np.abs(lasso.coef_)))

# Визуализируем зависимости
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(alpha_values, mse_lasso_values, marker='o')
plt.title('MSE в зависимости от alpha (Лассо)')
plt.xscale('log')
plt.xlabel('Alpha')
plt.ylabel('MSE')

plt.subplot(1, 2, 2)
plt.plot(alpha_values, max_coef_lasso_values, marker='o')
plt.title('Максимальный коэффициент в зависимости от alpha (Лассо)')
plt.xscale('log')
plt.xlabel('Alpha')
plt.ylabel('Максимальный коэффициент')

plt.show()
```

```python
#5 задание(SVR)
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler

X_svr = df[['TV', 'radio', 'newspaper']]
y_svr = df['sales']

# Нормализация признаков
scaler = StandardScaler()
X_svr_scaled = scaler.fit_transform(X_svr)

X_train_svr, X_test_svr, y_train_svr, y_test_svr = train_test_split(X_svr_scaled, y_svr, random_state=1)

# SVR с полиномиальным ядром
svr_poly = SVR(kernel='poly', C=100, epsilon=0.1)
svr_poly.fit(X_train_svr, y_train_svr)
y_pred_poly = svr_poly.predict(X_test_svr)
mse_poly = mean_squared_error(y_test_svr, y_pred_poly)
print("MSE (SVR Poly):", mse_poly)

# SVR с сигмоидальным ядром
svr_sigmoid = SVR(kernel='sigmoid', C=100, epsilon=0.1)
svr_sigmoid.fit(X_train_svr, y_train_svr)
y_pred_sigmoid = svr_sigmoid.predict(X_test_svr)
mse_sigmoid = mean_squared_error(y_test_svr, y_pred_sigmoid)
print("MSE (SVR Sigmoid):", mse_sigmoid)
```
